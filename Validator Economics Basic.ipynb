```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

np.set_printoptions(suppress=True)

TotalStakedSol = 300000000  # 2024 total staked sol
ProportionStakedSol_i = np.arange(0.0002, 1, 0.0002)
StakedSol_i = TotalStakedSol * ProportionStakedSol_i

EpY = 168 # Epochs per Year (as of August 2024)
C_v = 1.95 # voting costs (as of August 2024)
SlotsPerEpoch = 432000

TotalExpense = EpY * C_v                         # to include expected return of 1%: + StakedSol_i * 0.01
SlotsPerEpoch = 432000 
RewardPerSlot = 0.0156
TotalRevenue = SlotsPerEpoch * RewardPerSlot * EpY * ProportionStakedSol_i

Profit = TotalRevenue - TotalExpense
Yield = (Profit / StakedSol_i) * 100


#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# Plot the return of the SOL, proportionate to the amount staked (as a percentage)
# plotting
plt.title("Distribution vs. Proportion of Staked SOL") 
plt.xlabel("Proportion of Staked Sol") 
plt.ylabel("Annual Distribution (%)") 
plt.plot(ProportionStakedSol_i, Yield, color ="red")
```

```python

#plot 2:
ProportionStakedSol_i = np.arange(0.1, 1, 0.1)
TotalRevenue = SlotsPerEpoch * RewardPerSlot * EpY * ProportionStakedSol_i
Profit = TotalRevenue - TotalExpense

fig, ax = plt.subplots()
ax.plot(ProportionStakedSol_i, Profit, color="blue")
ax.set_title("Validator Profit from Block Rewards (SOL)")

# Set plain number formatting (no scientific notation)
ax.yaxis.set_major_formatter(ScalarFormatter(useOffset=False))
ax.ticklabel_format(style='plain', axis='y')

# Show the plot
plt.show()
```

```python
# plot 3 commission 

# TO BE SUSTAINABLE, SOLANA NEEDS 10X TRANSACTION THROUGHPUT (OR 10X FEES) SO THAT STAKERS CAN GET SOMETHING ABOVE A CONSEVERATIVE INVESTMENT RATE.
# When 1.5% becomes an unsustainable yield, there will be a race to compile on the validator that has the most stake 
# charge 0% commission as block rewards become more valuable... (block rewards will change as transactions increase decrease)
# there must be a point where max inflationary rewards = max block reward, and at this point this is where the proprtion stake will reach a steady state.
# for example @ current throughput, cut of inflation > max block reward (max block reward possible is 1.2 million @ 0.4% yield,thus charging @ 12m sol staked 8% comm on 7% inflation (around 600 thousand) is NO)
#   decision tree: charge 0% commission, aim for block reward
# @ 10x throughput, max block reward possible is 10m. @ inflation yield of 1.5%, with 12m staked sol, no such commission which matches max block reward.
#  decision tree: charge 0% commission, aim for block reward
# race to the top??
Validator_Commission = 0.10 # variable
SS_Inflation = 0.015
StakedSol_i = TotalStakedSol * ProportionStakedSol_i
Commisssion = StakedSol_i * SS_Inflation * Validator_Commission

fig, ax = plt.subplots()
ax.plot(ProportionStakedSol_i, Commisssion, color="blue")
ax.set_title("Validator Profit from 10% Commission (SOL)")

# Set plain number formatting (no scientific notation)
ax.yaxis.set_major_formatter(ScalarFormatter(useOffset=False))
ax.ticklabel_format(style='plain', axis='y')

# Show the plot
plt.show()
```
